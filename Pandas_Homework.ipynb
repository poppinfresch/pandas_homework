# Data Cleaning

In this section, you will need to read the CSV files into DataFrames and perform any necessary data cleaning steps. After cleaning, combine all DataFrames into a single DataFrame.

Files:

* `whale_returns.csv`: Contains returns of some famous "whale" investors' portfolios.

* `algo_returns.csv`: Contains returns from the in-house trading algorithms from Harold's company.

* `sp500_history.csv`: Contains historical closing prices of the S&P 500 Index.

## Whale Returns

Read the Whale Portfolio daily returns and clean the data

```python
# Import and Read in whale returns
whale_returns_path = Path("whale_returns.csv")
whale_df = pd.read_csv(whale_returns_path, index_col="Date", parse_dates=True, infer_datetime_format=True)
whale_df.head()
```

```python
#set date as index (sort ascending True)
whale_df = whale_df.sort_index(ascending=True)
whale_df.head()
```

```python
# Count nulls
whale_df.isnull().sum()
```

```python
# Drop nulls
whale_df = whale_df.dropna()
whale_df
```

## Algorithmic Daily Returns

Read the algorithmic daily returns and clean the data

```python
# Reading algorithmic returns
algo_returns_path = Path("algo_returns.csv")
algo_df = pd.read_csv(algo_returns_path, index_col="Date", parse_dates=True, infer_datetime_format=True)

algo_df.head()

```

```python
#set date as index
algo_df = algo_df.sort_index(ascending=True)
algo_df
```

```python
# Count nulls
algo_df.isnull().sum()

```

```python
# Drop nulls
algo_df = algo_df.dropna()
#then validate no null records remain
algo_df.isnull().sum()
```

```python
# check dataframe
algo_df.head(10)
```

## S&P 500 Returns

Read the S&P 500 historic closing prices and create a new daily returns DataFrame from the data. 

```python
# Reading S&P 500 Closing Prices
sp500_history_path = Path("sp500_history.csv")
sp500_df = pd.read_csv(sp500_history_path, index_col="Date", parse_dates=True, infer_datetime_format=True)
```

```python
# Check Data Types
df = pd.DataFrame(sp500_df)
print(df.dtypes)
```

```python
#Sort index by date
sp500_df = sp500_df.sort_index(ascending=True)
sp500_df.head()
```

```python
# Fix Data Types

#remove symbols first so we can change closing price values from str to float
sp500_df['Close'] = sp500_df['Close'].str.replace('$', '')

#change close column from str to float
sp500_df['Close'] = sp500_df['Close'].astype('float')
sp500_df['Close'].dtype

#check the data type 
df = pd.DataFrame(sp500_df)
print(df.dtypes)
```

```python
#rename 'close' column to its ticker name
sp500_df.columns.values[0] = "S&P 500"
sp500_df.head()
```

```python
# Calculate Daily Returns
sp500_df = sp500_df.pct_change()
sp500_df.head()
```

```python
#check nulls
sp500_df.isnull().sum()
```

```python
#drop nulls
sp500_df = sp500_df.dropna()
```

## Combine Whale, Algorithmic, and S&P 500 Returns

```python
# Join Whale Returns, Algorithmic Returns, and the S&P 500 Returns into a single DataFrame with columns for each portfolio's returns.
combined_returns = pd.concat([whale_df, algo_df, sp500_df], axis='columns', join='inner')
combined_returns = combined_returns.dropna()
combined_returns.head()
```

### Calculate and plot rolling `std` for all portfolios with 21-day window

```python
# Calculate the rolling standard deviation for all portfolios using a 21-day window
combined_rolling_std = combined_returns.rolling(window=21).std()

# Plot the rolling standard deviation
ax = combined_rolling_std.plot(figsize=(25,10), title="21 Day Rolling Standard Deviation of all Portfolios")
```

### Calculate and plot the correlation

```python
# Calculate the correlation
combined_correlation = combined_returns.corr()

# Display de correlation matrix
import seaborn as sns

sns.heatmap(combined_correlation, vmin=-1, vmax=1, annot=True)

```

### Calculate and Plot Beta for a chosen portfolio and the S&P 500

```python
# Calculate covariance of a single portfolio
covariance = combined_returns['S&P 500'].cov(combined_returns['Algo 1'])
print(f"Algo 1 covariance with S&P500 : {covariance}")

# Calculate variance of S&P 500
variance = combined_returns['S&P 500'].var()
print(f"Variance of S&P500 : {variance}")

# Computing beta
algo1_beta = covariance / variance
print(f"The Beta value of Algo 1 : {algo1_beta}")
```

```python
#calculate 60-day rolling covariance
rolling_covariance = combined_returns['S&P 500'].rolling(window=60).cov(combined_returns['Algo 2'])

#calculate 60-day rolling variance
rolling_variance = combined_returns['S&P 500'].rolling(window=60).var()

#calculate 60-day rolling beta
rolling_beta = rolling_covariance / rolling_variance

# Plot 60 day rolling beta between it and the S&P 500.
rolling_beta.plot(figsize=(20, 10), title='60-Day Rolling Beta Between Algo 1 and the S&P 500')
```

# Sharpe Ratios
In reality, investment managers and thier institutional investors look at the ratio of return-to-risk, and not just returns alone. After all, if you could invest in one of two portfolios, and each offered the same 10% return, yet one offered lower risk, you'd take that one, right?

### Using the daily returns, calculate and visualize the Sharpe ratios using a bar plot

```python
# Annualized Sharpe Ratios
sharpe_ratio = (combined_returns.mean()*252)/(combined_returns.std()*np.sqrt(252))
sharpe_ratio
```

```python
# Visualize the sharpe ratios as a bar plot
sharpe_ratio.plot(kind="bar", title="Sharpe Ratio of All Portfolios")
```

```python
# Reading data from 1st stock (Tezla): I used stocks I was interested in comparing to S&P500
tesla_csv = Path("tzla_portfolio - Sheet1.csv")
tzla_portfolio = pd.read_csv(tesla_csv, index_col="Date", parse_dates=True, infer_datetime_format=True)
```

```python
# Reading data from 2nd stock (AMZN): Amazon
amzn_csv = Path("amzn_portfolio - Sheet1.csv")
amzn_portfolio = pd.read_csv(amzn_csv, index_col="Date", parse_dates=True, infer_datetime_format=True)
```

```python
# Reading data from 3rd stock (AMC): AMC Entertainment
amc_csv = Path("amc_portfolio - Sheet1.csv")
amc_portfolio = pd.read_csv(amc_csv, index_col="Date", parse_dates=True, infer_datetime_format=True)
```

```python
# Combine all stocks in a single DataFrame
my_portfolio = pd.concat([amc_portfolio, amzn_portfolio, tzla_portfolio], axis='columns', join='inner')

# Reorganize portfolio data by having a column per symbol
my_portfolio.columns.values[0] = "AMC"
my_portfolio.columns.values[1] = "AMZN"
my_portfolio.columns.values[2] = "TZLA"
my_portfolio = my_portfolio.sort_index(ascending=True)
my_portfolio
```

```python
# Reset Date index for each dataframe before concatenating to ensure date index is the same
combined_returns.index = pd.to_datetime(combined_returns.index, errors='coerce').date
combined_returns.head()
```

```python
#reset date index so dataframes can align indexes
my_portfolio.index = pd.to_datetime(my_portfolio.index, errors='coerce').date
my_portfolio.head()
```

```python
# Calculate daily returns
# Drop NAs
my_returns = my_portfolio.pct_change().dropna()
my_returns
```

## Calculate the weighted returns for the portfolio assuming an equal number of shares for each stock

```python
# Set weights
weights = [1/3, 1/3, 1/3]

# Calculate portfolio return
weighted_returns = my_returns.dot(weights)

# Display sample data
weighted_returns.head()
```

## Join your portfolio returns to the DataFrame that contains all of the portfolio returns

```python
# Join your returns DataFrame to the original returns DataFrames 
joined_portfolio_returns = pd.concat([combined_returns, my_returns], axis="columns", join="inner")
joined_portfolio_returns
```

```python
# Only compare dates where return data exists for all the stocks (drop NaNs)
joined_portfolio_returns = joined_portfolio_returns.dropna()
joined_portfolio_returns
```

## Re-run the risk analysis with your portfolio to see how it compares to the others

### Calculate the Annualized Standard Deviation

```python
#calculate annualized standard deviation for your portfolio and compare it with the others
print(joined_portfolio_returns.std()*np.sqrt(252))
```

### Calculate and plot rolling `std` with 21-day window

```python
# Calculate rolling standard deviation
joined_rolling_std = joined_portfolio_returns.rolling(window=21).std()

# Plot rolling standard deviation
ax = joined_rolling_std.plot(figsize=(25,10), title="21 Day Rolling Standard Deviation of All Portfolios and the S&P 500")
```

### Calculate and plot the correlation

```python
# Calculate and plot the correlation
joined_portfolio_corr = joined_portfolio_returns.corr()

# plot with a heatmap to show correlation with all portfolios
sns.heatmap(joined_portfolio_corr, vmin=-1, vmax=1, annot=True)
```

   We can see from the heatmap, of the 3 stocks in my chosen portfolio, AMZN stock is more closely correlated to the S&P 500 than AMC or TZLA. Lets see how much volatility AMZN carries in comparison with the SP500.

### Calculate and Plot Rolling 60-day Beta for Your Portfolio compared to the S&P 500

```python
# Calculate and plot Beta

#calculate covariance of daily returns of my portfolio vs S&P500
my_covariance = joined_portfolio_returns["AMZN"].cov(joined_portfolio_returns["S&P 500"])
print(my_covariance)

#calculate variance of daily returns of my portfolio
my_variance = joined_portfolio_returns["AMZN"].var()
print(my_variance)

# Using variance and covariance of AMZN vs. SP500 we calculate beta for AMZN
amzn_beta = my_covariance / my_variance
print(f"Beta between AMZN and S&P500 : {amzn_beta}")
```

A beta value of 0.2823 for Amazon indicates that Amazon is approximately 60% less volatile than the S&P 500. A beta value greater than 1 would indicate more volatility than the compared stock/market.
